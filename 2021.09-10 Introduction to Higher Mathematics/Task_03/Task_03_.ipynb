## 1. Задание

Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)

1) Найдите их сумму. (на листочке) 

$\overrightarrow{a} = (10, 10, 10)$

$\overrightarrow{b} = (0, 0, -10)$

$\overrightarrow{a} + \overrightarrow{b} = (10+0, 10+0, 10+(-10)) = (10, 10, 0)$

2) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def calc_3d_vector_length(coord_start, coord_end):
    vec_len = ((coord_end[0] - coord_start[0]) ** 2 + 
               (coord_end[1] - coord_start[1]) ** 2 + 
               (coord_end[2] - coord_start[2]) ** 2) ** 0.5
    return vec_len
```

```python
a_start = (10, 10, 10) 
a_end = (0, 0, -10)
print(calc_3d_vector_length(a_start, a_end))
```

## 2. Задание (на листочке)

Почему прямые не кажутся перпендикулярными? (см.ролик)

<img src="Task_03_01.png" alt="Drawing" style="width: 400px;">

Прямые не кажутся перпендикулярными, потому что масштабы осей отличаются друг от друга. Попробую повторить код и изобразить линии на осях с одинаковым масштабом 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
rcParams["figure.figsize"] = 5, 5

x = np.linspace(-5, 5, 21)
y1 = 3 * x + 1
y2 = (-1 / 3) * x + 1

plt.plot(x, y1, label=f"y1")
plt.plot(x, y2, label=f"y2")

plt.xlabel("Переменная x")
plt.ylabel("Функция f(x)")

plt.legend(loc="upper left", frameon=False)

plt.axis([-5, 5, -4, 6])

plt.show()
```

На графике выше видно, что линии перпендикулярны.

## 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:
- окружности,	
- эллипса,
- гиперболы.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams, rcParamsDefault
```

```python
rcParams["figure.figsize"] = 5, 5

r = 5
x0 = 1
y0 = 1
x_min = x0 - r
x_max = x0 + r

step_lin = 0.5
semicircle_len = np.pi * r
n_points = int(semicircle_len / step_lin) + 1
steps_angle = np.linspace(0, np.pi, n_points)
x = np.array([x0 - r * np.cos(step_angle) for step_angle in steps_angle])

y_top = y0 + np.sqrt(r ** 2 - (x - x0) ** 2)
y_bottom = y0 - np.sqrt(r ** 2 - (x - x0) ** 2)

plt.plot(x, y_top, label=f"y", c="blue")
plt.plot(x, y_bottom, c="blue")

plt.title("Окружность")
plt.xlabel("Переменная x")
plt.ylabel("Функция f(x)")

plt.legend(loc="upper left", frameon=False)

# plt.axis([x_min-1, x_max+1, y0-r-1, y0+r+1])

plt.show()
```

```python
rcParams["figure.figsize"] = rcParamsDefault["figure.figsize"]

a = 10
b = 1
x0 = 10
y0 = 5

step_lin = 0.5
semicircle_len = np.pi * max(a, b)  # для аппроксимации здесь этого достаточно
n_points = int(semicircle_len / step_lin) + 1
steps_angle = np.linspace(0, np.pi, n_points)
x = np.array([x0 - a * np.cos(step_angle) for step_angle in steps_angle])

y_top = y0 + b * np.sqrt(1 - (x - x0) ** 2 / a ** 2)
y_bottom = y0 - b * np.sqrt(1 - (x - x0) ** 2 / a ** 2)

plt.plot(x, y_top, label=f"y", c="blue")
plt.plot(x, y_bottom, c="blue")

plt.title("Эллипс")
plt.xlabel("Переменная x")
plt.ylabel("Функция f(x)")

plt.legend(loc="upper left", frameon=False)

# x_min = x0 - a
# x_max = x0 + a
# plt.axis([x_min-1, x_max+1, y0-b-1, y0+b+1])

plt.show()
```

```python
rcParams["figure.figsize"] = rcParamsDefault["figure.figsize"]

a = 20
b = 4
x0 = -1
y0 = -2

# условие для `x` слева: x <= (-a + x0)
x_left = np.geomspace(-5*a, -1, 15)+1+(-a+x0)
# условие для `x` справа: x >= (a + x0)
x_right = np.geomspace(1, 5*a, 15)-1+(a+x0)

y_top_left = y0 + b * np.sqrt((x_left - x0) ** 2 / a ** 2 - 1)
y_top_right = y0 + b * np.sqrt((x_right - x0) ** 2 / a ** 2 - 1)
y_bottom_left = y0 - b * np.sqrt((x_left - x0) ** 2 / a ** 2 - 1)
y_bottom_right = y0 - b * np.sqrt((x_right - x0) ** 2 / a ** 2 - 1)

plt.plot(x_left, y_top_left, label=f"y", c="blue")
plt.plot(x_right, y_top_right, c="blue")
plt.plot(x_left, y_bottom_left, c="blue")
plt.plot(x_right, y_bottom_right, c="blue")

plt.title("Гипербола")
plt.xlabel("Переменная x")
plt.ylabel("Функция f(x)")

plt.legend(loc="center left", frameon=False)

# x_min = x0 - a
# x_max = x0 + a
# plt.axis([x_min-1, x_max+1, y0-b-1, y0+b+1])

plt.show()
```
