## 2. Задание (на листочке)
Даны три прямые 
$y=k_1 \bullet x+b_1, \\
y=k_2 \bullet x+b_2, \\
y=k_3 \bullet x+b_3 $ 
Как узнать, пересекаются они в одной точке или нет?

##### Решение

Если прямые пересекаются в одной точке, то значит в пространстве есть некая точка, координаты которой принадлежат всем линиям, которые пересекаются в этой точке. Нужно решить систему уравнений, у которой должен быть единственный корень (x, y) и этот корень должен удовлетворять всем трём уровнениям по отдельности:

$
\left \{ \begin{aligned} 
& y=k_1 \bullet x+b_1 \\
& y=k_2 \bullet x+b_2 \\
& y=k_3 \bullet x+b_3
\end{aligned} \right. 
\text{вычтем 2ое из 1го }
=> 
\left \{ \begin{aligned} 
& 0 = x \bullet (k_1 - k_2) + (b_1 - b_2) \\
& y=k_3 \bullet x+b_3
\end{aligned} \right.
=> 
\left \{ \begin{aligned} 
& x = \frac{b_2 - b_1}{k_1 - k_2} \\
& y=k_3 \bullet x+b_3
\end{aligned} \right.
$

Решая далее эту систему получаем корни (x, y), которые потом нужно проверить для каждого уравнения по отдельности.

## 3. Задание (в программе или на листочке)
На листе тетради «в линейку» (расстояние между линиями равно а) лежит игла (длиной b). Координаты нижней точки иглы (х,у), игла лежит под углом alfa. Пересекает ли игла линию или нет?

##### Решение

Пусть нижняя точка иглы имеет координаты (x1, y1), а верхняя - (x2, y2), тогда, чтобы игла пересекала линию тетради, должно удовлетвроряться неравенство |y2 - y1| > a. Также, если игла лежит под углом alfa к линии тетради, то sin(alfa) = |y2 - y1| / b или |y2 - y1| = b * sin(alfa), значит чтобы игла пересекала линию, должно удовлетвроряться неравенство 
b * sin(alfa) > a

Если нижняя точка иглы имеет координаты `(x, y)`, лежит под углом `alfa` к линиям тетради и имеет длину `b`, тогда верхняя точка будет иметь координаты: \
$ (x + b \bullet \cos \alpha, y + b \bullet \sin \alpha) $. \

Если расстояния между горизонтальными линиями тетради равно `a`, то уравнения линий можно записать как: \
$ f = a \bullet n, n \in N $.

Если игла пересекает линию, что нижняя её точка лежит ниже некоторой линии, а верхняя точка иглы лежит выше этой линии, тогда можно записать неравенство: \
$ y < a \bullet n < y + b \bullet \sin \alpha, n \in N $ \
или \
$
\left \{ \begin{aligned} 
& y < a \bullet n, n \in N \\
& y + b \bullet \sin \alpha > a \bullet n, n \in N
\end{aligned} \right.
$

Если наши данные удовлетворяют указанной системе неравенств, то игла пересекает как минимум одну линию.
