4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
rcParams["figure.figsize"] = 20, 5

x=np.linspace(-10, 10, 201)
k1=1
k2=0.5
y1 = np.cos(k1*x)
y2 = np.cos(k2*x)

plt.plot(x, y1, label=f"cos({k1} * x)")
plt.plot(x, y2, label=f"cos({k2} * x)")

plt.xlabel("Переменная x")
plt.ylabel("Функция f(x)")

plt.legend(loc="upper left", frameon=False)

plt.show()
```
